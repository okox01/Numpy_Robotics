```python
# temperature_analyzer
# Purpose: Learn how to compute statistics on data (mean, max, min) 
# â€” just like analyzing sensor readings from a robot or drone.

import numpy as np

temps=np.array([32,29,28,33,30,31,27]) # 7 days of temperature readings
mean_temp=np.mean(temps)
max_temp=np.max(temps)
min_temp=np.min(temps)

mean_temp,max_temp,min_temp
```

```python
# Brightness Adjustment
# Purpose: Practice manipulating images (arrays) 
# â€” foundational for computer vision in drones/robots. 
import numpy as np
import cv2
from matplotlib import pyplot as plt

# simulating a fake RGB image
img = np.random.randint(0,255,(200,300,3),dtype=np.uint8)
plt.imshow(img)
plt.title('Random RGB Image')

bright=img+240 # increase brightness by 240

bright=np.clip(bright,0,255) # ensure values stay within valid range

plt.imshow(bright) # display brightened image
plt.title('Brightened Image')
plt.axis('off')
```

```python
# Random Walk Simulation
# Purpose: Understand random processes â€” useful for simulating robot/drone movement.
import numpy as np
import matplotlib.pyplot as plt

steps = np.random.choice([-1, 1], size=100) # 100 random steps
pos = np.cumsum(steps) # cumulative sum to get position over time

plt.plot(pos) # plot the random walk in graph
plt.title("1D Random Walk")
plt.show()
```
