```python
# Mini Robotics Application: Drone Camera ROI
import numpy as np
import cv2
from matplotlib import pyplot as plt

#simulate a 480X740 RGB image (like a camera frame)
img=np.random.randint(0,255,(480,740,3),dtype=np.uint8)
# plt.imshow(img) - where plt.imshow() can be used to display images

# ROI - region of interest (200X300 area in the image)
roi=img[100:300,200:500]
# plt.imshow(roi) - where plt.imshow() can be used to display images with more zoom or focus

plt.imshow(roi)
plt.axis('off') # Hide axis
plt.title('Drone Camera ROI') # Title for the image
plt.show()
```

```python
# Add noise to the ROI
noise =np.random.randint(-180,180,roi.shape)  # Generate random noise
noisy_roi=np.clip(roi+noise,0,255) # Add noise and clip values to valid range

plt.imshow(noisy_roi)   # Display noisy ROI
plt.axis('off')
plt.title('Noisy Drone Camera ROI')
plt.show()
```
